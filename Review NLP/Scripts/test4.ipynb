```python
from transformers import pipeline
labels=[]

pipe = pipeline(model="facebook/bart-large-mnli")
for i in range(18294,21017):
  results = pipe(df["Review_Body"].iloc[i],candidate_labels=["camera", "battery", "display", "performance", "value for money"],)
  resultsdf=pd.DataFrame.from_dict(results)
  var = resultsdf["labels"].iloc[0]
  labels.append(var)
  print(i)
  #print(labels)
  #print(index)
my_list = list(zip(labels))
df_my_list_oneplus = pd.DataFrame(my_list,columns=['Label'])
df_my_list_oneplus.to_csv("Redmi11.csv")
```
